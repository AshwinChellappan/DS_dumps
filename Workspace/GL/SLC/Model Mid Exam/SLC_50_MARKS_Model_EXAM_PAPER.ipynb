```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
num_scaled=df_num.apply(lambda rec:rec-rec.mean()/rec.std(),axis=0)
```

```python
df_cat = df.select_dtypes(include="object")
df_cat = pd.get_dummies(df_cat, drop_first=True)
```

```python
X=pd.concat([df_num,df_cat],axis=1)
```

```python
X.head()
```

```python
x=X.drop('diagnosis_M',axis=1)
y=X['diagnosis_M']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
fpr1, tpr1, thresholds1 = roc_curve(y_pred,y_test)
plt.plot(fpr1, tpr1)
#plt.text( s = ('AUC Score for RFE model:',round(roc_auc_score(y_pred,y_test),4)))
plt.grid(True)
```
