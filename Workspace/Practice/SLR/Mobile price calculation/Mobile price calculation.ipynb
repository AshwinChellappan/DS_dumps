```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

```python
df_train['price_range'].value_counts()
```

```python
df_train.head()
```

```python
#Checking for train outlier
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1
df_train = df_train[~((df_train < (Q1 - 1.5 * IQR)) | (df_train > (Q3 + 1.5 * IQR))).any(axis=1)]
df_train.shape
```

```python
#Checking for test outlier
Q1 = df_test.quantile(0.25)
Q3 = df_test.quantile(0.75)
IQR = Q3 - Q1
df_test = df_test[~((df_test < (Q1 - 1.5 * IQR)) | (df_test > (Q3 + 1.5 * IQR))).any(axis=1)]
df_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()

df_train_scaled=minmax.fit_transform(df_train)
df_train_scaled=pd.DataFrame(data=df_train_scaled,columns=df_train.columns)
print(df_train_scaled.shape)


df_test_scaled=minmax.fit_transform(df_test)
df_test_scaled=pd.DataFrame(data=df_test_scaled,columns=df_test.columns)
print(df_test_scaled.shape)
```

```python
df_train_scaled.head()
```

```python
train_input_col=df_train_scaled.drop(columns=['price_range'],axis=1)
train_target=df_train_scaled['price_range']
test_input_col=df_test_scaled.drop(columns=['id'],axis=0)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
lin_model=LinearRegression().fit(train_input_col,train_target)
y_pred=lin_model.predict(test_input_col)
#mse_test=mean_squared_error(y_pred,y_test)
#print('Root Mean Squared Error:',round(np.sqrt(mse_test), 4))
r2_value_rfe = lin_model.score(train_input_col,train_target)
print('r-squared : ', r2_value_rfe)
#r2_value_rfe_test = lin_model.score(x_test, y_test)
#print('r-squared : ', r2_value_rfe_test)
```
