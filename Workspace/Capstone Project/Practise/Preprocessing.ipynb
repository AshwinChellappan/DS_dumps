```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import math
```

```python
df=pd.read_csv('data.csv')
```

```python
df.shape
```

```python
df['Bankrupt?'].value_counts()
```

```python
df.head()
```

```python
Q1 = df.quantile(0.01)
Q3 = df.quantile(0.99)
IQR = Q3 - Q1
df = df[~((df < (Q1-6*IQR)) | (df > (Q3 + 6*IQR))).any(axis = 1)]
```

```python
df.shape
```

```python
df['Bankrupt?'].value_counts()
```

```python
#df_scale=df.drop(columns='Bankrupt?')
```

```python
minmax=MinMaxScaler()
```

```python
df_scaled=minmax.fit_transform(df)
```

```python
df_scaled=pd.DataFrame(data=df_scaled,columns=df.columns)
```

```python
df_scaled.head()
```

```python
df_scaled.columns = df_scaled.columns.str.strip()
df_scaled.columns = df_scaled.columns.str.replace(" " ,"_")
df_scaled.rename(columns = {'Bankrupt?' :'Bankrupt' },inplace=True)
```

```python
df_scaled.shape
```

```python
df_scaled.head()
```

```python
df_scaled['Bankrupt']=df_scaled['Bankrupt'].astype('int64')
```

```python
df_scaled.head()
```

```python
df_scaled['Bankrupt'].value_counts()
```

```python
df_scaled.to_csv("datapreprocessed.csv")
```
