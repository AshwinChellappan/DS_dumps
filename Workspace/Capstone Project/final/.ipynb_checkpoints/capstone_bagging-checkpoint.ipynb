```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, StackingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_curve, roc_auc_score, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.decomposition import PCA
```

```python
LR = LogisticRegression(fit_intercept = True, max_iter = 500)
LR.fit(xtrain, ytrain)

y_pred_test = list(map(lambda x: 1 if x > 0.181 else 0, LR.predict_proba(xtest)[:, 1:]))
y_pred_train = list(map(lambda x: 1 if x > 0.181 else 0, LR.predict_proba(xtrain)[:, 1:]))

print(classification_report(ytest, y_pred_test))
print(classification_report(ytrain, y_pred_train))
```

```python
df = pd.DataFrame(confusion_matrix(ytest, y_pred_test), index = ["Actual 0", "Actual 1"], columns = ["Predicted 0", "Predicted 1"])
df
```

```python
Y_pred_LR = LR.predict_proba(X)[:, 1:]
```

```python
fpr, tpr, th = roc_curve(Y, LR.predict_proba(X)[:, 1:])
youden = pd.DataFrame({"TPR":np.round(tpr, 3), "FPR":np.round(fpr, 3), "Threshold":np.round(th,3)})
youden["Difference"] = youden.TPR - youden.FPR
youden.sort_values(by = ["Difference"], ascending = False, ignore_index = True).head()
```

```python
def cost(actual, predicted, cost_FN, cost_FP):
    cm = np.array(confusion_matrix(actual, predicted))
    return cm[1, 0]*cost_FN + cm[0, 1]*cost_FP

df_cut_off = pd.DataFrame(columns = ["cut_off", "cost"])
i = 0
for cut_off in range(10, 50):
    y_pred = list(map(lambda x: 1 if x > cut_off/100 else 0, Y_pred_LR))
    total_cost = cost(Y, y_pred, 2, 1)
    df_cut_off.loc[i] = [cut_off/100, total_cost]
    i+=1
df_cut_off.sort_values(by = ["cost"], ascending = True).head()
```
