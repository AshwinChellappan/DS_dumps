```python
boston=load_boston()
```

```python
bos=pd.DataFrame(boston.data,columns=boston.feature_names)
bos['price']=boston.target
bos.head()
```

```python
print(boston.DESCR)
```

```python
y=bos['price']
x=bos.drop('price',axis=1)
```

```python
import statsmodels.api as sm
xc=sm.add_constant(x)
model=sm.OLS(y,xc).fit()
model.summary()
```

```python
model.pvalues
```

```python
#Backward elimination Technique
#Algorithm to remove the non influential column affecting price
cols = list(xc.columns)
pmax=1

while len(cols)>0:
    X1 = xc[cols]
    model = sm.OLS(y, X1).fit()
    p = model.pvalues
    p = p.drop(labels = ['const'])  
    pmax = max(p)
    feature_max_p = p.idxmax() #Check and the highest value and return index of it
    if (pmax>0.05):
        cols.remove(feature_max_p)
        print('Removed',feature_max_p,'prob',pmax)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```
